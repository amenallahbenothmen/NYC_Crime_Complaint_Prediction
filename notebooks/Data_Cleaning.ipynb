```python
df['Start_Datetime'] = pd.to_datetime(
    df['Start_Date'] + ' ' + df['Start_Time'], 
    format="%m/%d/%Y %H:%M:%S", 
    errors='coerce'  # Handle invalid dates gracefully
)
df['End_Datetime'] = pd.to_datetime(
    df['End_Date'] + ' ' + df['End_Time'], 
    format="%m/%d/%Y %H:%M:%S", 
    errors='coerce'
)
```

```python
conditions = [
    df['Offense_Description'].isin(prop),
    df['Offense_Description'].isin(sexual),
    df['Offense_Description'].isin(drug_alch),
    df['Offense_Description'].isin(personal),
    df['Offense_Description'].isin(admin)
]
categories = ['PROPERTY_RELATED', 'SEXUAL_RELATED', 'DRUGS/ALCOHOL_RELATED', 'PERSONAL_RELATED', 'ADMINISTRATIVE_RELATED']
df['Offense_Description'] = np.select(conditions, categories, default='OTHER')
```

```python
import matplotlib.pyplot as plt

filtered_data = df[
    (df['Borough'] != 'Unknown') &
    (df['Offense_Description'] != 'OTHER')
]
crime_borough_distribution = (
    filtered_data.groupby(['Borough', 'Offense_Description'])
    .size()
    .unstack(fill_value=0)
)

crime_borough_percentages = crime_borough_distribution.div(crime_borough_distribution.sum(axis=1), axis=0) * 100

fig, ax = plt.subplots(figsize=(16, 10))  
crime_borough_percentages.plot(
    kind='bar',
    figsize=(16, 10),
    ax=ax,
    width=0.8,
    colormap='tab10'
)


for i, borough in enumerate(crime_borough_percentages.index):
    for j, offense_type in enumerate(crime_borough_percentages.columns):
        percentage = crime_borough_percentages.loc[borough, offense_type]
        if percentage > 0:  
            ax.text(i + (j - 2) * 0.18, percentage + 1, f"{percentage:.1f}%", ha='center', fontsize=11, color='black')

ax.set_title('Crime Type Distribution by Borough (Percentage)', fontsize=20, pad=20)
ax.set_xlabel('Borough', fontsize=16, labelpad=10)
ax.set_ylabel('Percentage (%)', fontsize=16, labelpad=10)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.legend(title='Offense Description', title_fontsize=14, fontsize=12, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```
